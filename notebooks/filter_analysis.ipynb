```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

from processing import preprocess_new, preprocess
from get_data import get_data_by_id_activity, get_participant_ids
from files import load_raw_data
```

```python
acquisition_folderpath = 'Aquisicao'
id_participants = get_participant_ids(acquisition_folderpath)
data_old_preprocess, data_raw = get_data_by_id_activity(acquisition_folderpath, alternative_preprocessing=False)
data_new_preprocess, _ = get_data_by_id_activity(acquisition_folderpath, alternative_preprocessing=True)
```

```python

id = "EPE2"
activity = "SGB"
sensor = "MAG"

sensor_dict = {"MAG": "mag", "PZT": "pzt"}
sampling_freq =  100


fig = make_subplots(specs=[[{"secondary_y": True}]])
# plot sensor and airflow signals
fig.add_trace(go.Scatter(x=np.linspace(1, len(data_old_preprocess[id][activity][sensor_dict[sensor]]),len(data_old_preprocess[id][activity][sensor_dict[sensor]]))/sampling_freq, 
                            y=data_old_preprocess[id][activity][sensor_dict[sensor]], 
                            name=sensor))
fig.add_trace(go.Scatter(x=np.linspace(1, len(data_old_preprocess[id][activity]['airflow']),len(data_old_preprocess[id][activity]['airflow']))/sampling_freq,
                            y=data_old_preprocess[id][activity]['airflow'], 
                        name="Airflow", line={"color":"white"}), secondary_y=True)
fig.show()

fig = make_subplots(specs=[[{"secondary_y": True}]])
# plot sensor and airflow signals
fig.add_trace(go.Scatter(x=np.linspace(1, len(data_new_preprocess[id][activity][sensor_dict[sensor]]),len(data_new_preprocess[id][activity][sensor_dict[sensor]]))/sampling_freq, 
                            y=data_new_preprocess[id][activity][sensor_dict[sensor]], 
                            name=sensor))
fig.add_trace(go.Scatter(x=np.linspace(1, len(data_new_preprocess[id][activity]['airflow']),len(data_new_preprocess[id][activity]['airflow']))/sampling_freq,
                            y=data_new_preprocess[id][activity]['airflow'], 
                        name="Airflow", line={"color":"white"}), secondary_y=True)
fig.show()

```
